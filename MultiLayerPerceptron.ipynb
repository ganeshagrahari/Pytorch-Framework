```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('Churn_Modelling.csv')
```

```python
df.head()
```

```python
df.drop(columns=['RowNumber','CustomerId','Surname','Geography','Gender',], inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
#Diving in to features and labels
x = df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
#Scaling
ss = StandardScaler()
x=pd.DataFrame(ss.fit_transform(x),columns = x.columns)
```

```python
#train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x
```

```python
#creating An ANN
import tensorflow
```

```python
from keras.layers import Dense
from keras.models import Sequential
```

```python
ann  = Sequential()
```

```python
x.shape
```

```python
ann.add(Dense(6,input_dim = 8,activation ="relu")) #input layer
ann.add(Dense(4,activation ="relu")) #hidden layer 1
ann.add(Dense(2,activation ="relu")) #hidden layer 2
ann.add(Dense(1,activation ="sigmoid")) # in the last layer we are changing the af cuz now there is no hidden layer
```

```python
ann.compile(optimizer = "adam", loss ="binary_crossentropy", metrics = ["accuracy"])
```

```python
ann.fit(x_train,y_train,batch_size=100,epochs= 50)
```

```python
from sklearn.metrics import accuracy_score
```

```python
prdtest=ann.predict(x_test)
```

```python
prdtest_data = []
for i in prdtest:
  if(i[0]>0.5):
    prdtest_data.append(1)
  else:
      prdtest_data.append(0)
```

```python
prdtrain=ann.predict(x_train)
```

```python
prdtrain_data = []
for i in prdtrain:
  if(i[0]>0.5):
    prdtrain_data.append(1)
  else:
      prdtrain_data.append(0)
```

```python
accuracy_score(y_test,prdtest_data) #1-> actual_value 2-> predicted_value
```

```python
accuracy_score(y_train,prdtrain_data)
```
